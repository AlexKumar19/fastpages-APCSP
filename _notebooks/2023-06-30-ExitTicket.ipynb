```python
%%html

<!DOCTYPE html>
<html>
<head>
    <style>
        /* CSS styling */
        .container {
            display: flex;
            justify-content: center;
            align-items: center;
            height: 100vh;
        }
        
        .box {
            width: 100px;
            height: 100px;
            background-color: blue;
            margin: 10px;
            color: white;
            text-align: center;
            line-height: 100px;
            font-size: 24px;
        }
    </style>
</head>
<body>
    <div class="container">
        <div class="box">1</div>
        <div class="box">2</div>
        <div class="box">3</div>
        <div class="box">4</div>
        <div class="box">5</div>
    </div>

    <script>
        // JavaScript logic

        // Function to shuffle an array using Fisher-Yates algorithm
        function shuffleArray(array) {
            for (let i = array.length - 1; i > 0; i--) {
                let j = Math.floor(Math.random() * (i + 1));
                [array[i], array[j]] = [array[j], array[i]];
            }
            return array;
        }

        // Function to animate the shuffling of boxes
        function animateShuffle() {
            const boxes = Array.from(document.getElementsByClassName('box'));
            const shuffledBoxes = shuffleArray(boxes);

            shuffledBoxes.forEach((box, index) => {
                setTimeout(() => {
                    box.style.order = index + 1;
                }, (index + 1) * 500);
            });
        }

        // Function to animate the sorting of boxes using bubble sort algorithm
        function animateSort() {
            const boxes = Array.from(document.getElementsByClassName('box'));
            let sorted = false;
            let pass = 0;

            function swap(a, b) {
                const temp = boxes[a].style.order;
                boxes[a].style.order = boxes[b].style.order;
                boxes[b].style.order = temp;
            }

            function bubbleSort() {
                sorted = true;
                for (let i = 0; i < boxes.length - pass - 1; i++) {
                    if (parseInt(boxes[i].style.order) > parseInt(boxes[i + 1].style.order)) {
                        swap(i, i + 1);
                        sorted = false;
                    }
                }
                pass++;

                if (!sorted) {
                    setTimeout(bubbleSort, 500);
                }
            }

            bubbleSort();
        }

        // Triggering the animations
        animateShuffle();
        setTimeout(animateSort, 5000);
    </script>
</body>
</html>
```
